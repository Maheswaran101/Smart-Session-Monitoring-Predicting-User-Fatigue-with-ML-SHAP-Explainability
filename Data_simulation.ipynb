## Creating A Dataset

```python
# Importing Data Scince Libraries
import pandas as pd
import numpy as np
# Importing Visualization Libraires
import matplotlib.pyplot as plt
import seaborn as sns
# Importing Machine Learning Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Defining users & Sessions

```python

num_user = 500

session_per_user = np.random.randint(2, 6, size=num_user)

data = []
```

## Generate Features

```python
for user_id in range(1, num_user+1):
    for session in range(session_per_user[user_id-1]):
        session_duration = np.random.randint(10, 180)
        decision_count = np.random.randint(5, 120)
        undo_count = np.random.randint(0, 20)
        error_rate = round(np.random.uniform(0, 0.4), 2)
        break_taken = np.random.choice([0, 1], p=[0.6, 0.4])
        time_of_day = np.random.choice(['Morning', 'Afternoon', 'Evening'])

        # Fatigue score logic
        fatigue_score = (session_duration*0.3) + (decision_count*0.4) + (undo_count*0.4) + (error_rate*10)
        if break_taken ==1:
            fatigue_score -=15

        # Convert to category
        if fatigue_score < 30:
            fatigue_level = "Low"
        elif fatigue_score < 60:
            fatigue_level = "Medium"
        else:
            fatigue_level = "High"  

        # Append to list
        data.append([user_id, session, session_duration, decision_count, undo_count, error_rate, break_taken, time_of_day, fatigue_score, fatigue_level])
```

## Create Dataframe

```python
columns = ['user_id', 'session_id', 'session_duration', 'decision_count', 'undo_count', 'error_rate', 'break_taken', 'time_of_day', 'fatigue_score', 'fatigue_level']   

df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
df.to_csv('data.csv', index=False)

print("Data saved to 'data.csv'")
```
